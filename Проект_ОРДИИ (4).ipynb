##Визуализация данных

```python
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Распределение возраста")
plt.xlabel("Возраст")
plt.ylabel("Количество")
plt.show()
```

```python
avg_age_purchase = df.groupby("Age")["Purchase Amount (USD)"].mean()
plt.plot(avg_age_purchase)
plt.title("Средняя сумма покупки в зависимости от возраста")
plt.xlabel("Возраст")
plt.ylabel("Средняя сумма покупки (USD)")
plt.show()
```

```python
df["Gender"].value_counts().plot(kind="bar")
plt.title("Распределение по полу")
plt.xlabel("Пол")
plt.ylabel("Количество клиентов")
plt.show()
```

```python
sns.boxplot(x="Payment Method", y="Purchase Amount (USD)", data=df)
plt.title("Boxplot суммы покупок по способу оплаты")
plt.xlabel("Способ оплаты")
plt.ylabel("Сумма покупки (USD)")
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

##Группировки и агрегации

```python
avg_by_gender = df.groupby("Gender")["Purchase Amount (USD)"].agg(["mean"])
avg_by_gender
#Среднее суммы покупок по полу
```

```python
count_payment = df.groupby("Payment Method")["Customer ID"].count()
count_payment
#Количество клиентов по способу оплаты
```

```python
sum_by_rating = df.groupby("Review Rating")["Purchase Amount (USD)"].sum()
sum_by_rating
#Сумма покупок по рейтингу отзывов
```
